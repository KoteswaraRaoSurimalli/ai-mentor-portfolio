## What Big-O Actually Is

Think of it this way: you write a function to search for a name in a list of 1000 students. It works fine. But what happens when the college has 1 lakh students? Does your code slow down a little, a lot, or not at all?

**Big-O notation is just your answer to that question — mathematically.**

It describes how your algorithm's *time or space* grows as the input size (we call it **n**) grows. It's not about seconds on a clock. It's about *rate of growth*.

---

## The Ones You Must Know Cold

**O(1) — Constant Time**
No matter how big the input, it takes the same time.
```python
arr[5]  # accessing an index directly
```
Doesn't matter if the array has 10 or 10 million elements. One step. Done.

---

**O(log n) — Logarithmic**
Input doubles, but your work only increases by 1 step. This is **Binary Search**.
```python
# Searching in a sorted array — you halve the problem each time
```
1000 elements? ~10 steps. 1 million elements? ~20 steps. That's the power of log n.

---

**O(n) — Linear**
You visit every element once. Classic **linear search**.
```python
for item in arr:
    if item == target: return item
```
Double the input → double the time. Fair trade.

---

**O(n log n) — The sorting zone**
This is where **Merge Sort and Quick Sort** live. Almost every efficient sorting algorithm lands here. Interviewers love asking *why* sorting is n log n — know the merge sort recursion tree explanation.

---

**O(n²) — Quadratic**
Nested loops. **Bubble Sort, Selection Sort, Insertion Sort.**
```python
for i in arr:
    for j in arr:   # every element checks every other
```
1000 elements → 1 million operations. This is the one TCS/Infosys MCQs love to trap you on.

---

**O(2ⁿ) / O(n!) — Avoid in interviews**
Recursive Fibonacci without memoization, brute-force permutations. If your solution lands here, the interviewer is waiting for you to *optimise it*. That's often the whole point of the question.

---

## The Trick Interviewers Actually Use

They give you code like this and ask: *"What's the complexity?"*

```python
for i in range(n):          # O(n)
    for j in range(n):      # O(n)
        print(i, j)
```
**Answer: O(n²)**

Then they modify it slightly:
```python
for i in range(n):          # O(n)
    print(i)
for j in range(n):          # O(n)
    print(j)
```
**Answer: O(n) — not O(n²).** Sequential loops *add*, nested loops *multiply*.

This single distinction comes up in almost every TCS Digital and Infosys SP paper.

---

## The "Drop the Constants" Rule

Big-O ignores constants and lower-order terms.

- O(2n) → just say **O(n)**
- O(n² + n) → just say **O(n²)**
- O(500) → just say **O(1)**

Why? Because at massive scale, constants become irrelevant. The *shape* of growth is what matters.

---

## How to Answer in an Interview

When they ask *"what's the time complexity of your solution?"* — don't just blurt out O(n). Say:

> *"The outer loop runs n times, and the inner operation is O(1), so the overall time complexity is O(n). Space complexity is O(1) since I'm not using any extra data structure."*

That one sentence — covering **both time and space, with a brief why** — is what separates a 7.5 CGPA candidate from the one who gets the offer.

**Question-1**
```markdown
What is recursion?
```
**OUTPUT**
Recursion is when a function calls itself as part of its own definition, breaking a problem down into smaller versions of the same problem until it reaches a **base case** that stops the recursion.

**The two essential parts:**
1. **Base case** — the condition that stops the recursion
2. **Recursive case** — where the function calls itself with a simpler input

**Classic example — factorial:**

```python
def factorial(n):
    if n == 0:          # base case
        return 1
    return n * factorial(n - 1)  # recursive case

factorial(4)
# → 4 * factorial(3)
# → 4 * 3 * factorial(2)
# → 4 * 3 * 2 * factorial(1)
# → 4 * 3 * 2 * 1 * factorial(0)
# → 4 * 3 * 2 * 1 * 1  =  24
```

**A useful mental model:** Think of recursion like Russian nesting dolls — each doll contains a smaller version of itself, until you reach the smallest one that doesn't open.

**When recursion fits naturally:**
- Tree/graph traversal
- Divide-and-conquer algorithms (merge sort, quicksort)
- Problems defined recursively (Fibonacci, fractals)
- Parsing nested structures (JSON, HTML)

**One caveat:** Each recursive call adds a frame to the call stack. Too many calls without hitting the base case causes a *stack overflow*. For deep recursion, an iterative approach or tail-call optimization may be preferable.

**Question-2**
```markdown
What is a stack?
```
A stack is a data structure that follows **LIFO** — Last In, First Out. Think of it like a stack of plates: you can only add or remove from the top.

**Core operations:**
- `push` — add an item to the top
- `pop` — remove the item from the top
- `peek` — look at the top item without removing it

**Simple example:**

```python
stack = []

stack.append(1)   # push → [1]
stack.append(2)   # push → [1, 2]
stack.append(3)   # push → [1, 2, 3]

stack.pop()       # → 3, stack is now [1, 2]
stack.pop()       # → 2, stack is now [1]
```

**Where stacks show up:**
- **Call stack** — how your program tracks function calls (directly relevant to recursion!)
- **Undo/redo** — text editors push changes onto a stack
- **Browser history** — the back button pops the last page
- **Expression parsing** — balancing parentheses, evaluating math expressions
- **Depth-first search** — graph traversal

**Connection to your last question:** When a recursive function calls itself, each call gets *pushed* onto the call stack. When it hits the base case, the calls start *popping* off — which is exactly why too-deep recursion causes a *stack overflow* (the stack runs out of space).


**Question-3**
```markdown
What is Big-O notation?
```
**OUTPUT**
Big-O notation describes how an algorithm's **time or space requirements grow** as the input size grows. It answers: *"How does this scale?"*

It focuses on the **worst case** and **dominant term** — ignoring constants and smaller factors.

**Common complexities, fastest to slowest:**

| Notation | Name | Example |
|---|---|---|
| O(1) | Constant | Array lookup by index |
| O(log n) | Logarithmic | Binary search |
| O(n) | Linear | Looping through a list |
| O(n log n) | Linearithmic | Merge sort |
| O(n²) | Quadratic | Nested loops |
| O(2ⁿ) | Exponential | Recursive Fibonacci |
| O(n!) | Factorial | Brute-force permutations |

**Quick examples:**

```python
# O(1) — always one step
def get_first(lst):
    return lst[0]

# O(n) — grows with input
def find_item(lst, target):
    for item in lst:
        if item == target:
            return item

# O(n²) — nested loops over same input
def has_duplicates(lst):
    for i in range(len(lst)):
        for j in range(len(lst)):
            if i != j and lst[i] == lst[j]:
                return True
```

**The "ignore constants" rule:**
- `O(2n)` → simplifies to `O(n)`
- `O(n² + n)` → simplifies to `O(n²)`
- `O(500)` → simplifies to `O(1)`

The idea is that at large scale, the *shape* of growth matters more than the exact multiplier.

**Connection to your earlier questions:** The recursive Fibonacci is a classic O(2ⁿ) algorithm — each call spawns two more calls, and the call stack grows explosively. That's why memoization (caching results) is used to bring it down to O(n).

**1. Definition & Formal Meaning**
Big-O notation is a mathematical way to describe how the runtime or space usage of an algorithm *scales* with input size — not what the exact time is, but how it *grows*. Formally, f(n) = O(g(n)) means there exist constants c and n₀ such that f(n) ≤ c·g(n) for all n ≥ n₀. For interviews, the key intuition is this: you're answering "if I double the input size, what happens to the cost?" Big-O gives you the ceiling — the worst your algorithm can do as input grows large. Interviewers don't want exact runtimes; they want you to reason about growth behavior.

**2. Worst-Case vs. Average-Case vs. Best-Case**
Big-O is almost always talking about *worst-case* unless explicitly stated otherwise. Quick Sort is O(n log n) on average but O(n²) in the worst case — and in an interview, you're expected to know both. Best-case analysis (Ω) is rarely useful in practice because you can't guarantee ideal inputs. When a company asks "what's the time complexity of your solution?", they mean worst-case by default. Mixing these up is one of the most common mistakes freshers make in placement interviews, so always clarify which case you're discussing.

**3. Dropping Constants & Lower-Order Terms**
When analyzing complexity, you drop constants and lower-order terms because Big-O describes behavior at large n, where they become irrelevant. O(3n² + 5n + 7) becomes O(n²) because for very large n, the n² term completely dominates. This is not an approximation — it's mathematically rigorous. In interviews, writing O(2n) instead of O(n) signals that you don't fully understand the notation. The practical takeaway: focus on the fastest-growing term and strip everything else. This habit also helps you quickly estimate complexity during live coding rounds without getting bogged down in arithmetic.

**4. Common Complexity Classes**
You must instinctively recognize the standard complexity hierarchy: O(1) < O(log n) < O(n) < O(n log n) < O(n²) < O(2ⁿ) < O(n!). Each maps to real patterns — hash table lookups are O(1), binary search is O(log n), merge sort is O(n log n), and naive subset enumeration is O(2ⁿ). In placement interviews, recognizing which class your solution falls into — and whether a better class is achievable — is critical. If your brute force is O(n²) and the interviewer nudges you to optimize, they usually expect O(n log n) or O(n). Knowing this hierarchy tells you *how much room for improvement exists*.

**5. Space Complexity vs. Time Complexity**
Big-O applies equally to memory and runtime, and interviewers frequently ask about both. A recursive solution might be O(n) time but also O(n) space due to the call stack — which matters in memory-constrained systems. An iterative version of the same logic might be O(1) space. In placement interviews, especially at product-based companies, trading space for time (like using a hash map to reduce O(n²) to O(n)) is a core optimization pattern you're expected to suggest. Always analyze both dimensions in your answer — it signals maturity and practical awareness beyond just solving the problem.
